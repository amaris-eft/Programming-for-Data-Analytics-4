5. Implement logistical regression model to the banking dataset and explore the relationship between 'age','education','default' ,'duration','pdays','previous' and 'y'. Explain your results. (30 points)

- Data source: https://raw.githubusercontent.com/madmashup/targeted-marketing-predictive-engine/master/banking.csv


Variables:
1. age (numeric)
2. education (categorical: basic.4y, basic.6y, basic.9y, high.school, illiterate, professional.course, university.degree, unknown)
3. default: has credit in default? (categorical: no, yes, unknown)
4. duration: last contact duration, in seconds (numeric).
5. pdays: number of days that passed by after the client was last contacted from a previous campaign (numeric; 999 means client was not previously contacted)
6. previous: number of contacts performed before this campaign and for this client (numeric)

y: has the client subscribed a term deposit? (binary: 1, means Yes, 0 means No)

```python
#5
import pandas as pd
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import classification_report, confusion_matrix

df2 = pd.read_csv('https://raw.githubusercontent.com/madmashup/targeted-marketing-predictive-engine/master/banking.csv')
df2.head()
```

```python
df2.shape #tuple of array dimensions
df2.dropna(inplace = True)
df2.shape 
```

```python
import seaborn as sns
df2.groupby('y').mean()
sns.histplot(x = 'education', data= df2, bins = 30)
df2.corr()
```

```python
import matplotlib.pyplot as plt
df2.groupby(['y','education']).mean()

#here is the age frequency plot: age (numeric) 
df2.groupby('y').age.plot(kind='kde', stacked=True)
plt.legend()
plt.xlabel('Age')
plt.show()

#here is the education frequency plot: education (categorical: basic.4y, basic.6y, basic.9y, high.school, illiterate, professional.course, university.degree, unknown)
t1 = pd.crosstab(df2.education, df2.y)
t1.div(t1.sum(1), axis = 0).plot(kind = 'bar', stacked = True)

#here is the default plot: default: has credit in default? (categorical: no, yes, unknown)
t2 = pd.crosstab(df2.default, df2.y)
t2.div(t2.sum(1), axis = 0).plot(kind = 'bar', stacked = True)

#here are the past days default plot: number of days that passed by after the client was last contacted from a previous campaign (numeric; 999 means client was not previously contacted)
t3 = pd.crosstab(df2.pdays, df2.y)
t3.div(t3.sum(1), axis = 0).plot(kind = 'bar', stacked = True)

#here is the previous plot: number of contacts performed before this campaign and for this client (numeric)
t4 = pd.crosstab(df2.previous, df2.y)
t4.div(t4.sum(1), axis = 0).plot(kind = 'bar', stacked = True)
```
